```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
sick = pd.read_excel(r'C:\Users\USER\Desktop\프로젝트\csv\그래프용\환자연령비율\J18.xls', encoding = 'UTF-8')
sick
```

```python
j18 = pd.pivot_table(sick, index=['성별구분'], columns=["연령구분10세"], values=['총환자수'], aggfunc=sum, fill_value=0)
j18
```

```python
labels = ['0_9세', '10_19세', '20_29세', '30_39세', '40_49세', '50_59세', '60_69세', '70_79세', '80세이상']
sizes = [789237, 75247, 39978, 79713, 84122, 114636, 134729, 145109, 127249]
explode = (0, 0, 0, 0, 0, 0, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 

plt.title ('폐렴 환자 연령비율')
plt.tight_layout()
plt.show()
```

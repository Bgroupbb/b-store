```python
import matplotlib.pyplot as plt

%matplotlib inline

import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 

import pandas as pd
import numpy as np
```

```python
daegupo = pd.read_excel(r'C:\Users\USER\Desktop\프로젝트\csv\그래프용\대구\daegupo.xls')
daegupo
```

```python
width = 0.3
N = 9
ind = np.arange(N)

data1 = daegupo['PM10']
data2 = daegupo['PM2.5']

plt.figure(figsize=(7, 6))
p1 = plt.bar(ind+0.00, data1, width, color='y')
p2 = plt.bar(ind+0.33, data2, width, color='r')

plt.xlabel('미세먼지배출종류')
plt.ylabel('배출량')
plt.title('대구시 미세먼지 배출(2015년 기준)')
plt.xticks(ind+width/2., ('기타', '도로', '비도로', '비산업', '생물성', '생산', '에너지', '제조업', '폐기물'))
plt.legend((p1[0], p2[0]), ('PM10', 'PM2.5'))
plt.show
```

```python
labels = ['기타', '도로', '비도로', '비산업', '생물성', '생산공정', '에너지', '제조', '폐기물' ]
sizes3 = [11.145, 394.542, 200.922, 52.275, 125.813, 1.255, 33.114, 680.370, 7.629 ]
explode = (0.3, 0, 0, 0, 0, 0.3, 0.1, 0, 0.1)
fig3, ax3 = plt.subplots()
ax3.pie(sizes3, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax3.axis('equal') 

plt.title ('대구 오염원 비중(PM10)')
plt.tight_layout()
plt.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
daegu = pd.read_excel(r'C:\Users\USER\Desktop\프로젝트\csv\그래프용\대구\구별업체.xls')
```

```python
N = 7
ind = np.arange(N)

my_plot = daegu.plot(kind='bar',stacked=True, title="구별",figsize=(11, 7))
plt.ylabel('업체 갯수')
plt.xticks(ind, ('서구', '북구', '동구', '수성구', '달서구', '남구', '달성군', '중구' ))
plt.show()
```

```python
labels = ['서구', '북구', '동구', '수성구', '달서구', '남구', '달성군', '중구']
sizes = [80, 219, 192, 120, 213, 50, 153, 39]
explode = (0, 0, 0, 0, 0, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 

plt.title ('구별 경로당 비율')
plt.tight_layout()
plt.show()
```

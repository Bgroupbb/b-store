```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.request import urlopen
```

```python
url_base = "http://192.168.0.82"
url_sub = "/oldpeople.php"
url =url_base + url_sub

html = urlopen(url)
soup = BeautifulSoup(html, "html.parser")

soup
```

```python
사업장명 = []
주소 = []


list_soup = soup.find_all('tr')

for item in list_soup:
    사업장명.append(item.find(class_='사업장명').get_text())
    주소.append(item.find(class_='주소').get_text())
```

```python
data = {'사업장명':사업장명, '주소':주소}
df = pd.DataFrame(data)
df.head()
```
